```python
import seaborn as sns # Import the seaborn library and give it the alias 'sns'
```

```python
num_cols_eda = df_eda.columns

fig, ax = plt.subplots(4, 4, figsize=(12, 12))
ax = ax.flatten()

for i, col in enumerate(num_cols_eda):
    sns.histplot(df_eda[col],kde=True,stat='density',bins=30,ax=ax[i])

for i in range(len(num_cols_eda),16):
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(4, 4, figsize=(12, 12))
ax = ax.flatten()

for i, col in enumerate(num_cols_eda):
    df.boxplot(col, ax=ax[i])

for i in range(len(num_cols_eda),16):
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```
